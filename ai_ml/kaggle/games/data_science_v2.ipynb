**<간단 산술 통계>를 위한 기능**

* **info()**: 데이터의 간단한 정보를 살펴봅니다.
* **describe()**: 데이터의 산술 통계 정보를 살펴봅니다.
* **isnull()**: 데이터가 누락되었는지 여부를 확인합니다.
* **sum()**: 데이터를 합산합니다. (Excel의 sum)
* **value_counts()**: 데이터의 종류별 갯수를 확인합니다. (사과: 2개, 배: 3개)

**<고급 산술 통계>를 위한 기능**
* **groupby (혹은 pivot_table)**: 엑셀의 피벗테이블과 동일합니다. 연도별, 성별 통계랑과 같이 특정 column으로 데이터를 뽑아낼 때 활용합니다.

**<조건 필터링>을 위한 기능**
* **loc**: 조건문에 대한 필터를 걸어 데이터를 확인합니다.
* **& 와 | 연산자**: & (and) 조건은 모두 만족하는 경우. | (or)조건은 하나의 조건이라도 만족하는 경우를 말합니다.


**<데이터 전처리>를 위한 기능**
* **replace()**: 데이터를 다른 값으로 일괄 바꿔줄 때 활용합니다.
* **fillna()**: 빈 값을 채워줄 때 활용합니다. fill na 입니다. 여기서 na는 NaN (Not a Number)를 지칭하며, 숫자 뿐만아니라 비어 있는 값을 통틀어 지칭합니다.
* **map()**: 데이터를 매핑해줍니다. 가령 사과=1, 배=2라고 매핑해준다고 한다면, 사과가 들어 있는 데이터는 모두 1로, 배가 들어 있는 데이터는 모두 2로 변경해 줍니다.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint

# 1. 게임 데이터셋 로드
df = pd.read_csv('Free-to-play-games.csv')

# 2. 데이터 전처리
tfidf = TfidfVectorizer(stop_words='english')
# short_description	genre	platform	publisher	developer

df['combined_features'] = df['title'] + ' ' + df['genre'] + ' ' + df['short_description']
tfidf_matrix = tfidf.fit_transform(df['combined_features'])

# 3. 유사도 계산
def get_similar_games(user_games, top_n=20):
    user_games_tfidf = tfidf.transform(user_games)
    user_games_tfidf = user_games_tfidf.T
    cosine_sim = cosine_similarity(tfidf_matrix, user_games_tfidf.T)
    
    # 각 게임별 유사도 점수 계산
    sim_scores = [sum(score) for score in cosine_sim]
    
    # 유사도 점수 기반 상위 N개 게임 선정
    indices = sorted(range(len(sim_scores)), key=lambda i: sim_scores[i], reverse=True)[:top_n]
    
    # 추천 게임 정보 반환
    return df.iloc[indices]

# 4. 사용자 입력 받기
user_games = ['KurtzPel', 'Genshin Impact', 'Overwatch 2', 'Kritika: Zero', 'Dragon Nest', 'Digimon Masters Online', 'Elsword', 'League of Legends', 'Seal Online', 'MapleStory']

# 5. 추천 게임 출력
recommended_games = get_similar_games(user_games)
recommended_games = [game for game in recommended_games['title'] if game not in user_games]

pprint(recommended_games)
```

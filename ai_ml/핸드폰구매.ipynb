```python
import pandas as pd

# 매장 목록
# columns = ['쿠팡_자급', '퍼스_공시']
columns = ['쿠팡_자급', '알뜰_자급', '퍼스_공시', '퍼스_선택']


# 요금제 1개월~24개월
# S24 울트라 기준
rows = ['단말기값'] + [f'요금제{i}' for i in range(1, 25)] + ['회사지원'] + ['선물혜택']

# 지원항목
회사_지원 = [-39000*12]
선택약정선물 = [-300000]

# 가격테이블
쿠팡_자급 = [1510450] + [45000]*24 + 회사_지원 
알뜰_자급 = [1510450] + [26000]*24 + [0]
퍼스_공시 = [618400+38720] + [109000]*6 + [69000]*18 + 회사_지원 
퍼스_선택 = [1498400] + [87150]*3 + [51750]*21 + 회사_지원 + 선택약정선물

# Create the DataFrame
df = pd.DataFrame([쿠팡_자급, 알뜰_자급, 퍼스_공시, 퍼스_선택], columns=rows, index=columns).T
df = df.fillna(0).astype(int)
print(df)
print("====총합====",df.sum(axis=0), sep="\n")
print("====참고====")
print("1. 25000원 요금제부터 알뜰폰이 이득")
print("2. 공시든 선택이든 걍 자급제가 쌈. 다만 비싼 요금제를 잠깐 누릴 수 있음. 근데 바꾸는게 더 귀찮...")
print("3. 선택약정시 갤탭a9준다고 해서 계산해보니 선택약정이 제일 싸긴함")
```
